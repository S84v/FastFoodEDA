```python
fig, axes = plt.subplots(2, 4, figsize=(25, 10))
fig.suptitle("Graphs sorted by restaurants (Calories x Fat)", size = "xx-large")
conc = "Fat content increases linearly in each restaurant with calories"
fig.text(0.5, 0, conc, ha='center', size = 'x-large')
```

```python
sns.regplot(ax=axes[0,0], x='calories', y='cal_fat', data = df[df['restaurant']=='Mcdonalds'], color = "red").set(title="Mcdonalds")
sns.regplot(ax=axes[0,1], x='calories', y='cal_fat', data = df[df['restaurant']=='Chick Fil-A'], color = "red").set(title="Chick Fil-A")
sns.regplot(ax=axes[0,2], x='calories', y='cal_fat', data = df[df['restaurant']=='Sonic'], color = "gray").set(title="Sonic")
sns.regplot(ax=axes[0,3], x='calories', y='cal_fat', data = df[df['restaurant']=='Arbys'], color = "black").set(title="Arbys")
sns.regplot(ax=axes[1,0], x='calories', y='cal_fat', data = df[df['restaurant']=='Burger King'], color = "yellow").set(title="Burger King")
sns.regplot(ax=axes[1,1], x='calories', y='cal_fat', data = df[df['restaurant']=='Dairy Queen'], color = "orange").set(title="Dairy Queen")
sns.regplot(ax=axes[1,2], x='calories', y='cal_fat', data = df[df['restaurant']=='Subway'], color = "green").set(title="Subway")
sns.regplot(ax=axes[1,3], x='calories', y='cal_fat', data = df[df['restaurant']=='Taco Bell'], color = "magenta").set(title="Taco Bell")
```

```python
df_p  = df.loc[df['item'].str.contains('pork', case=False)]
df_b  = df.loc[df['item'].str.contains('bacon', case=False)]
df_h  = df.loc[df['item'].str.contains('ham', case=False)]
df_br = df.loc[df['item'].str.contains('brisket', case=False)]
df_bf = df.loc[df['item'].str.contains('beef', case=False)]
df_c = df.loc[df['item'].str.contains('chicken', case = False)]
```

```python
fig, ax = plt.subplots(2,3, figsize=(15,10))

sns.regplot(ax=ax[0,0], data=df_p, x='calories', y='protein', color='c').set_title(label='Pork', color='c')
sns.regplot(ax=ax[0,1], data=df_b, x='calories', y='protein',color = 'y').set_title(label='Bacon', color='y')
sns.regplot(ax=ax[0,2], data=df_h, x='calories', y='protein', color = 'm').set_title(label='Ham', color='m')
sns.regplot(ax=ax[1,0], data=df_br, x='calories', y='protein', color = 'k').set_title(label='Brisket', color='k')
sns.regplot(ax=ax[1,1], data=df_bf, x='calories', y='protein', color = 'r').set_title(label='Beef', color='r')
sns.regplot(ax=ax[1,2], data=df_c, x='calories', y='protein', color = 'b').set_title(label='Chicken', color='b')
```
